```python
import os

# output들을 저장해놓은 폴더 노트북을 위치해 놓고, 모든 output들을 불러옵니다.
path = "./"
file_list = os.listdir(path)
file_list_csv = [file for file in file_list if file.endswith(".csv")]

print("file_list: {}".format(file_list_csv))
```

```python
import pandas as pd

output_list = []

for file in file_list_csv:
    output_list.append(pd.read_csv(file))
```

```python
output_list[0]
```

```python
df = pd.DataFrame()
for idx, output in enumerate(output_list):
    df[file_list_csv[idx]] = output['prediction']
```

```python
df
```

```python
corrM = df.corr()
corrM
```

```python
cluster_1 = [146]
cluster_2 = [111, 102, 104, 107, 113, 103, '092', 100]
cluster_3 = [133, 123, 131]
cluster_4 = [139, 140, 142, 132, 141]
cluster_5 = ['085']
cluster_6 = ['087', '086', '089', '050', '067']

col_name_list = ['cluster_1', 'cluster_2', 'cluster_3', 'cluster_4', 'cluster_5', 'cluster_6']
cluster_list = [cluster_1, cluster_2, cluster_3, cluster_4, cluster_5, cluster_6]

for idx in range(len(col_name_list)):
    df[col_name_list[idx]] = 0
    
    for col in cluster_list[idx]:
        cluster_name = 'output_{}.csv'.format(col)
#         if len(cluster_list[idx]) == 1:
#             df[col_name_list[idx]] += df[cluster_name] 
#         else:
#             df[col_name_list[idx]] += df[cluster_name] ** 2
        df[col_name_list[idx]] += df[cluster_name] ** 1.8

    df[col_name_list[idx]] /= len(cluster_list[idx])

df
```

```python
df_clusters = df[col_name_list]
df_cluster_avg = df_clusters.mean(axis=1)
```
